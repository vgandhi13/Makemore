log(a*b*c) = log(a) + log(b) + log(c)

```python
log_likelihood = 0.0
n = 0
for w in words[:3]:
  chs = ['<.>'] + list(w) + ['<.>'] #start and end token
  for ch1, ch2 in zip(chs,chs[1:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]
    prob = P[ix1, ix2]
    logprob = torch.log(prob)
    log_likelihood += logprob
    n+=1
    print(f'{ch1}{ch2}: {prob:.4f} {logprob:.4f}')

print('Log likelihood is ', log_likelihood) #highest it can go to 0, when alk probs lower, this will become more and more negative
#But we dont actually like this because we use loss function, and it has semantics that low is good because we try ot minimize the loss so we need to invert this. and hence negative log likelihood

nll = -log_likelihood
print(nll)# so this is a very good loss functionn, because the lowest it can go is 0, the higher it is the worse off predictions are. for conveniance people make it average instead of just sum. the lower it is the better model is.
print(nll/n)#average neg lle
```

mle: summarize these probs into one number. it is product of all the probaboloitos of the actual correct words. it tells us the prob of entire dataset assigned by the model we have trained. that is measure of quality. so product should be as high as possible when training the model.the product will be tiny number due to decimals, hence we use log of the probability.

goal: maximize likelihood of the data wrt model parameters(statistical modeling)
equivalent to maximizing the log likelihood (because it is monotonic)
equivalent to minimizing the negative log likelihood
equivalent to minimizing the average log likelihood

model smoothing: needed as when some prob 0, log prb is inf. you can add 1 to logits and it will ensure no zero in prob matrix p. nothing will be inf for log prb and nothingbreaks

```python
nlls = torch.zeros(5)
for i in range(5):
  # i-th bigram
  x = xs[i].item()
  y = ys[i].item()
  print('----')
  print(f'{x}:{itos[x]}{y}:{itos[y]}')
  print(x,y, probs[i])
  p = probs[i, y]
  print('true prob', p.item())
  logp = torch.log(p)
  print('log likelihood', logp.item())
  nll = -logp
  nlls[i] = nll
  print('negative log likelihood', nll.item())
print('----')

print('average negative log likelihood, ie. loss = ', nlls.mean().item())
```

what every element of w.grad tells us is the influence of that weight on loss function. for example (0,0), bcoz grad is positive, it means it has a positive influence on the loss. slightly nudging w[0,0] and adding a small h would increase the loss mildly because the gradient is positive.

```python
for i in range(100):
  #forward pass
  xenc = F.one_hot(xs, num_classes=27).float()
  logits = xenc @ W
  counts = logits.exp()
  probs = counts / counts.sum(1, keepdims = True)

  loss= -probs[torch.arange(228146), ys].log().mean() + 0.01*(W**2).mean() #neg log likelihhood
  print(loss)
  W.grad = None #None more efficient than setting to 0, pytorech understands lack of gradient
  loss.backward()
  W.data += -50 * W.grad
```

insights:
the w that we get after trained is equivalent to the counts in the table for the simple bigram model. And hence, the loss for both of them is same when w is idetical

model smoothing and regulariztion

if W intialized with 0, then logits all 0, and exp all 1, and then probs all uniform. so when w are equal to each other or equal to 0, the probs come uniform. so trying to incentivize w to be near 0 is equivalent to label smoothing. the more incentivize this in loss function, the more smooth distribuition we ahchieve.
regularization: we can augment loss function to have small component called regularization loss. This adds like a spring force to push W to be zero. W wants to be zero and probs want to be uniform, but they also simultaneously want to match your probabilities as indicated by the data. the strength of this regukarization is like controlling the N+1 previously

#Makemore 2 - MLP


Impelement MLP model to predict next char- Follows Bengio's paper - a neurol probabilistic model. THEY Build a word language model, but we stick to characters

Every 17000 word embedded to 30 dim space feature . Begining they are intialized randomly, then we tune them using backprop, they will move around the space, words that have similar meaning will end up in similar part of the high dimensional space. They maximized log likelihood. Input layer has 30000 neurons for each word. Hidden layer neurons are hyper param. Ouptut layer has 17000 neurons for each word.

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
g = torch.Generator().manual_seed(2147483647)
```

```python
words = open('names.txt', 'r').read().splitlines()
words[:8]
```

```python
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i, s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s, i in stoi.items()}
itos
```

```python
#build the dataset

block_size = 3
X, Y = [], []
for w in words[:5]:
  print(w)
  context = [0] * block_size
  for ch in w + '.':
    ix = stoi[ch]
    X.append(context)
    Y.append(ix)
    print(''.join(itos[i] for i in context), '--->', itos[ix])
    context = context[1:] + [ix] # crop and append

X = torch.tensor(X)
Y = torch.tensor(Y)
```

```python
parameters=[C,w1,b1,w2, b2]
for p in parameters:
  p.requires_grad=True
```

```python
for _ in range(1000):
  emb = C[X] #(32,3,2)
  h = torch.tanh(emb.view(-1,6)@w1 + b1) #(32,100)
  logits = h @w2+b2
  loss = F.cross_entropy(logits,Y)
  print(loss.item())
  for p in parameters:
    p.grad=None
  loss.backward()
  for p in parameters:
    p.data+=-0.1*p.grad
```
