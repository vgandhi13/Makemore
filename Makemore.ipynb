```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
%matplotlib inline
g = torch.Generator().manual_seed(2147483647)
```

```python
words= open('names.txt','r').read().splitlines()
words[:8]
```

```python
len(words)
```

```python
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i, s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s, i in stoi.items()}
block_size = 3
vocab_size = len(itos)
itos
```

```python

def build_dataset(words):
  X, Y =[], []
  for w in words:
    context = [0] * block_size
    for ch in w + '.':
      ix = stoi[ch]
      X.append(context)
      Y.append(ix)
      context = context[1:] + [ix]
  X = torch.tensor(X)
  Y = torch.tensor(Y)
  print(X.shape, Y.shape)
  return X, Y

import random
random.seed(42)
random.shuffle(words)

n1 = int(0.8*len(words))
n2 = int(0.9*len(words))
Xtr, Ytr = build_dataset(words[:n1])
Xdev, Ydev = build_dataset(words[n1:n2])
Xte, Yte = build_dataset(words[n2:])
```

```python
#MLP revisited
g = torch.Generator().manual_seed(2147483647) # seeds a generator to get similar results

n_embd = 10 #the dimensionality of the character embedding vectors
n_hidden = 200 # the number of neurons in the hidden layer of the MLP
block_size = 3

C = torch.randn(vocab_size, n_embd, generator=g) #27X10
W1 = torch.randn(block_size * n_embd, n_hidden, generator=g) * (5/3)/((block_size * n_embd)**0.5) #on left kaiming init for tanh #* 0.2
#b1 = torch.randn(n_hidden, generator=g) * 0.02  #can help a lil bit more than 0 for optimization
W2 = torch.randn(n_hidden, vocab_size, generator = g) * 0.01 # cant make this 0 because
b2 = torch.randn(vocab_size, generator=g) * 0

bngain = torch.ones((1, n_hidden))
bnbias = torch.zeros((1, n_hidden))
bnmean_running = torch.zeros((1, n_hidden)) #not part of optimization, kept on side running for forward pass on single example during testing
bnstd_running = torch.ones((1, n_hidden)) #not part of optimization, kept on side running for forward pass on single example during testing

parameters = [C, W1, W2, b2, bngain, bnbias] #5 groups of params b1,
print(sum(p.nelement() for p in parameters)) #number of individual parameters in total, if a weight matrix is of size (100, 27), p.nelement() would return 2700 (100 * 27).
for p in parameters:
  p.requires_grad = True#so we can comptue grads for backprop
```

```python
max_steps = 200000
batch_size = 32
lossi = []

for i in range(max_steps):
  ix = torch.randint(0, Xtr.shape[0], (batch_size,), generator = g) # creates a tensor of shape 32, with values between 0 and nelems in xtr
  Xb, Yb = Xtr[ix], Ytr[ix] #batch X,Y, size [32, 3]

  #forward pass
  emb = C[Xb] #[32, 3, 10], C had [27, 10] - Inpute layer
  embcat = emb.view(emb.shape[0], -1) #[32, 30] #-1 is a special value that tells PyTorch to infer the size of the remaining dimensions. In this case, it will flatten the last two dimensions (3 * 10 = 30).
  hpreact = embcat @ W1 #+ b1 #[32,30] X [30, 200] + [200,] = [32,200] - hidden layer preactivation value - There are 200 neurons in this hidden layer
  bnmeani = hpreact.mean(0, keepdim=True)
  bnstdi = hpreact.std(0, keepdim=True)
  hpreact = bngain * (hpreact - bnmeani)/bnstdi + bnbias

  with torch.no_grad():
    bnmean_running = 0.9999 * bnmean_running + 0.001 * bnmeani
    bnstd_running = 0.9999 * bnstd_running + 0.001 * bnstdi

  h = torch.tanh(hpreact) #[32,200] - hidden layer
  logits = h @ W2 + b2 # [32,200] X [200, 27] - output layer - raw scores/log counts
  loss = F.cross_entropy(logits, Yb) # exponentiate + normalization + log + mean - cross entrpy loss is equivalent to negative log likelihood when softmax is applied

  for p in parameters:
    p.grad = None
  loss.backward()

  lr = 0.1 if i < 100000 else 0.01 #initially higher learning rate, when start converging then learning rate is lower so we dont over shoot
  for p in parameters:
    p.data += -lr * p.grad # - lr because we want to go in opposite direction of the gradient, gradient points in the direction of increase of loss

  #track stats - print once in a while
  if i % 10000 == 0:
    print(f'{i:7d}/{max_steps:7d}: {loss.item():.4f}')
    lossi.append(loss.log10().item())
  # break
```

Issues:

1. at the very first iteration we were recordng a loss of 27.85. This rapidly came down to 1 or2. This is way to high in the begining and we know our initialization is wrong. In training neural nets we almost always have an estimate of what loss should be during initialization it depends on loss fn and prblm setup. Here at inititialization, 27 chars can come next and we have no reason to believe any characters to be much more likely than others. So, we'd expect that prob dist that comes out will be uniform dist assigning about equal prob to all chars. So prob of any char, = 1/27
loss = -torch.tensor(1/27).log() = 3.2958
This is much lower than 27. What is happening right now, that neural net creating prob dist very messed up and it is just very very wrong. We basically want logits to be same for all when neural net is initialized to get correct loss. One way is to make them all closer to 0 by scaling the params

In the previous case where initial high, first few steps of optimization is just squashing the logits and then it is rearranging the logits. Hence, th loss here will be slight worse as less cycles spent/more time in optimizing instead first few iterations were just spent squashing down the weights.

Batch normalization - we have hidden states hpreact. We dont want these preactivation states to be too small because then tan h is not doing anything. But we also dont want it to be too large because then tan h is saturate and values close to 1. We want them to be roughly gaussian atleast at init. Batch normalization says take hidden states and normalize them to be unit gaussian.We will use a formula standardis the weights. And after this, firing rate of every single neuron on these 32 samples will be unit gaussian. We want this to be unit gaussian only during initialization, but we dont want these to be forced to be gaussian always. wE like the neural net to move this distribution around to potentially make it more diffused, more sharp, to make some tanh neurons more trigger happy,or less trigger happy. So we'd like this distribution to move around and we would like backpropagation to tell us how to move this distribution around. Scale and shift helps with this.
Because these are initializaed to 1 and 0, at init, each neurons firing value will be exact unit gaussian, regardless of value of hpreact . No matter what dist of hpreact is coming in, on coming out, it will be approximate unit gaussian, which is exactly what we want atleast for the initial stages. During, optimization we will be able to backpropagate bngain and bnbias and change them so network is given full ability to do this whatever it wants internally.

We will be adding these batch normalization throughout the neural net and then this controls the scale of the activations throughout the neural net. It doesnt require us to care about the activation distribution for all diff types of building blocks and significantly stabilizes training. It comes at a cost however. IT HAS some sort of a regularizing effect , it stabilizerd training, and people have tried to remove it but it has been hard because it works quite well because of this regularizing effect becasue it is quite good at controlling these activations and their distributions

```python
class Linear: #Torch.nn has a lot of layers one of which is linear
  def __init__(self, fan_in, fan_out, bias=True):
    self.weight = torch.randn((fan_in, fan_out), generator=g) / fan_in**0.5 #kaiming init
    self.bias = torch.zeros(fan_out) if bias else  None

  def __call__(self, x):
    self.out = x @ self.weight
    if self.bias is not None:
      self.out += self.bias
    return self.out

  def parameters(self):
    return [self.weight] + ([] if self.bias is None else [self.bias])
```

```python
class BatchNorm1d:
  def __init__(self, dim, eps=1e-5, momentum=0.1):
    self.eps = eps
    self.momentum = momentum
    self.training = True
    #parameters (trained with backprop)
    self.gamma = torch.ones(dim)
    self.beta = torch.zeros(dim)
    #buffers(trainied with a running momentum update)
    self.running_mean = torch.zero(dim)
    self.running_var = torch.ones(dim)

  def __call__(self, x):
    if self.training:
      xmean = x.mean(0, keepdim=True) #batch mean
      xvar = x.var(0, keepdim=True) #batch variance
    else:
      xmean = self.running_mean
      xvar = self.running_var
    xhat = (x - xmean) / torch.sqrt(xvar + self.eps) #normalize to unit variance
    self.out = self.gamma * xhat + self.beta
    #update the buffers
    if self.training:
      with torch.no_grad():
        self.running_mean = (1- self.momentum) * self.running_mean + self .momentum *xmean
        self.running_var = (1- self.momentum) * self.running_var + self.momentum *xvar
    return self.out

  def parameters(self):
    return [self.gamma, self.beta]
```
